```python
import seaborn as sns
diamonds=sns.load_dataset("diamonds")
df=diamonds.copy()
df.head()
```

```python
df.dtypes
```

```python
df.info()
```

```python
df.describe().T
```

```python
df["cut"].value_counts().plot.barh();
## kalite karşılaştırması yapıldı
```

```python
df["clarity"].value_counts().plot.barh(color="red");
## veri setinde olan elmasların beraklığını kıyaslandı
```

```python
df["color"].value_counts().plot.barh(color="green");
## veri setinde elması renklerini kıyaslandı
```

```python
df.cut.head()
```

```python
sns.barplot(x="cut",y=df.cut.index, data=df);
```

```python
sns.catplot(x="cut",y="price",data =df);
## kalite ile değer arasındaki grafik durumuna baktık 
```

```python
sns.barplot(x="cut",y="price",hue="color",data=df);
## burda calitenin paraya dağılımının renge göre dağılımı
```

```python
sns.distplot(df.price,bins=1000,kde=False);
# veride bins değeri ölçeklendirmeni sağlar , değerin hangi alanda yoğulduğunu gösterir
```

```python
sns.distplot(df.price);
```

```python
sns.kdeplot(df.price,shade=True,color="red");
```

```python
(sns.FacetGrid(df,
               
               hue="cut",
              height=5,
              xlim =(0,10000))
 .map(sns.kdeplot, "price",shade=True)
 .add_legend()
);
## grafik üzerine eklenen boyutları göstermemizi sağlar  hue : boyut eklemek için kulanılır 
```

```python
sns.catplot(x="cut",y="price",hue="color",kind="point",data=df);
```
